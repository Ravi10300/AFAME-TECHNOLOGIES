```python
import pandas as pd


file_path = "HR Data.csv" 
data = pd.read_csv(file_path)

data.head()
```

```python
columns_to_remove = ['EmployeeCount', 'StandardHours', 'Over18', 'EmployeeNumber']
data_cleaned = data.drop(columns=columns_to_remove, axis=1)

data_cleaned.head()
```

```python

columns_renamed = {
    'DistanceFromHome': 'Distance_Home',
    'EducationField': 'Education_Field',
    'JobSatisfaction': 'Job_Satisfaction',
    'JobInvolvement': 'Job_Involvement',
    'YearsWithCurrManager': 'Years_With_Manager',
}
data_cleaned.rename(columns=columns_renamed, inplace=True)

data_cleaned.head()
```

```python
data_cleaned.drop_duplicates(inplace=True)


print(f"Number of rows after removing duplicates: {data_cleaned.shape[0]}")
```

```python
data_cleaned['Attrition'] = data_cleaned['Attrition'].str.strip().str.capitalize()
data_cleaned['BusinessTravel'] = data_cleaned['BusinessTravel'].str.replace('_', ' ').str.capitalize()

data_cleaned[['Attrition', 'BusinessTravel']].head()
```

```python
data_cleaned.dropna(inplace=True)

print(f"Number of missing values: {data_cleaned.isnull().sum().sum()}")
```
